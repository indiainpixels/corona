```python
def ffy(yr):
    title = "https://www.imdb.com/title/tt2467372/episodes?season="+yr
    data = requests.get(title)
    soup = BeautifulSoup(data.text.strip())
    voo = []
    for div in soup.find_all("div",{"class","list_item"}):
        name = div.find("a")
        sc = div.find("span",{"class","ipl-rating-star__rating"})
        print(sc)
        voo.append({"name":name["title"], "score":sc.text})
    return voo
```

```python
toRet = []
for y in range(1,9):
    toRet.append(ffy(str(y)))
```
