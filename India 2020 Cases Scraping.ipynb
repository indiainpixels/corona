```python
url = "https://api.covid19india.org/v4/timeseries.json"
import requests, json, moment
```

```python
initdate = moment.date("2020-01-30")
```

```python
initdate
```

```python
newdate = initdate.clone().add(days=335)
```

```python
newdate.format("YYYY-MM-DD")
```

```python
data = requests.get(url)
```

```python
jso = data.json()
```

```python
jso['KL']
```

```python
alldates = {}
```

```python
for day in range(0,len(jso['KL']["dates"].keys())):
    newdate = initdate.clone().add(days=day)
    nd = newdate.format("YYYY-MM-DD")
    for state in jso:
        try:
            delconf = jso[state]['dates'][nd]['delta']['confirmed']
        except:
            delconf = 0
        if day not in alldates:
            alldates[day] = {}
        alldates[day][state] = delconf
```

```python
alldates
```

```python
with open("2020.json",'w') as f:
    json.dump(alldates,f)
```
