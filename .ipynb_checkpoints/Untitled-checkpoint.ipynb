```python
def formatize(table):
    del table["Lat"]
    del table["Long"]
    del table["Province/State"]
    return table.groupby("Country/Region").sum()
```

```python
import copy
def drow(LL):
    L=copy.deepcopy(LL[-15:-1])
    Z = [y - x for x,y in zip(L,L[1:])]
    V = [y / x for x,y in zip(Z,Z[1:]) if x!=0]
    if len(V)>0:
        return sum(V)/len(V)
    else:
        return 0
```

```python
drow([1,2,3,30,60,120,240,480])
```
